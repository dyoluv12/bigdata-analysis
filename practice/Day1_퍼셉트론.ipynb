# **퍼셉트론**

- ### 다수의 신호를 입력으로 받아 하나의 신호를 출력한다
- ### 전류와 비슷하게 신호도 흐름을 만들고 정보를 앞으로 전달한다
### -> 전류와 달리 퍼셉트론 신호는 흐른다 / 안 흐른다 (1 또는 0)으로 두 가지 값을 가진다

#### 1 : 신호가 흐른다

#### 0 : 신호가 흐르지 않는다

## 단층 퍼셉트론 : 입력이 2개인 퍼셉트론

- 입력 신호 x1, x2
- 출력 신호 y
- x1의 가중치 w1, x2의 가중치 w2

#### 입력 신호가 노드에 보내질 때 각각의 고유한 가중치가 곱해진다

#### 가중치는 각 신호가 결과에 주는 영향력을 조절한다

#### 가중치가 클수록 해당 신호의 중요도가 높다

#### 뉴런에서 보낸 신호의 총 합이 theta(임계값)를 넘을 때에만 1을 출력한다

## 1. 퍼셉트론 구현

x1과 x2를 인수로 받는 AND 함수 구현

```python
def AND(x1, x2):
  w1, w2, theta = 0.5, 0.5, 0.7
  tmp = x1 * w1 + x2 * w2
  if tmp <= theta:
    return 0
  else:
    return 1
```

```python
print(AND(0,0))
print(AND(0,1))
print(AND(1,0))
print(AND(1,1))
```

### weight와 bias를 이용한 AND 게이트

```python
def AND(x1, x2):
  x = np.array([x1, x2])
  w = np.array([0.5, 0.5])
  b = -0.7
  tmp = np.sum(w*x) + b
  if tmp <= 0:
    return 0
  else:
    return 1
```

```python
print(AND(0, 0))
print(AND(0, 1))
print(AND(1, 0))
print(AND(1, 1))
```

### OR 게이트

```python
def OR(x1, x2):
  x = np.array([x1, x2])
  w = np.array([0.5, 0.5])
  b = -0.2
  tmp = np.sum(w*x) + b
  if tmp <= 0:
    return 0
  else:
    return 1
```

```python
print(OR(0,0))
print(OR(0, 1))
print(OR(1, 0))
print(OR(1, 1))
```

### NAND 게이트

```python
def NAND(x1, x2):
  x = np.array([x1, x2])
  w = np.array([-0.5, -0.5])
  b = 0.7
  tmp = np.sum(w*x) + b
  if tmp <= 0:
    return 0
  else:
    return 1
```

### XOR 게이트 - AND, OR, NAND

```python
def XOR(x1, x2):
  s1 = NAND(x1, x2)
  s2 = OR(x1, x2)
  y = AND(s1, s2)
  return y
```

```python
print(XOR(0,0))
print(XOR(0,1))
print(XOR(1,0))
print(XOR(1,1))
```
